```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import metrics
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
df0 = pd.read_csv('data/data_0.csv')
df1 = pd.read_csv('data/data_1.csv')
```

```python
#Obtener las metricas que utilizaremos para ver la precision del modelo
def compare_models(dic_models, X, y, cv_folds):
    results = []
    kf = KFold(n_splits=cv_folds, shuffle=True)

    for name, model in dic_models.items():
        scores_r2 = cross_val_score(model, X, y, cv=kf, scoring='r2')
        scores_rmse = cross_val_score(model, X, y, cv=kf, scoring='neg_root_mean_squared_error')

        results.append({
            'MODEL': name,
            'R2 MEAN': scores_r2.mean(),
            'R2 STD': scores_r2.std(),
            'RMSE MEAN': scores_rmse.mean()
        })
    
    return pd.DataFrame(results).sort_values(by='R2 MEAN', ascending=False)
    
```

```python
dic_models = {
    'Linear simple': linear_model.LinearRegression(),
    'Linear L1': linear_model.Lasso(),
    'Linear L2': linear_model.Ridge(),
    'Polynomial 2': make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression()),
    'Polynomial 3': make_pipeline(PolynomialFeatures(3), linear_model.LinearRegression()),
    'Polynomial 4': make_pipeline(PolynomialFeatures(4), linear_model.LinearRegression()),
    'Polynomial 5': make_pipeline(PolynomialFeatures(5), linear_model.LinearRegression()),
    'Polynomial 6': make_pipeline(PolynomialFeatures(6), linear_model.LinearRegression()),
    'Polynomial 7': make_pipeline(PolynomialFeatures(7), linear_model.LinearRegression()),
    'Polynomial 8': make_pipeline(PolynomialFeatures(8), linear_model.LinearRegression()),
    'Polynomial 9': make_pipeline(PolynomialFeatures(9), linear_model.LinearRegression()),
    'Polynomial 10': make_pipeline(PolynomialFeatures(10), linear_model.LinearRegression()),

}

metrics_table = compare_models(dic_models, np.array(df_time).reshape(-1,1), df_variation, 10)
```
